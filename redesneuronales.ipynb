ESTE CODIGO DESCOMPRIME EL ARCHIVO ZIP, LUEGO EL ZIP DENTRO DE EL Y ACCEDE AL CONTENIDO DE LA CARPETA TRAIN SIN DESCARGAR NADA AL REPOSITORIO

```python
import zipfile
import os

def listar_archivos_zip(ruta_zip):
    nombres_archivos = []  # Lista para almacenar los nombres de los archivos
    with zipfile.ZipFile(ruta_zip, 'r') as zip_ref:
        for archivo in zip_ref.namelist():
            if archivo.endswith('train.zip'):
                with zip_ref.open(archivo, 'r') as train_zip_file:
                    nombres_archivos.extend(listar_archivos_train_zip(train_zip_file))  # Agregar nombres de archivos

    return nombres_archivos

def listar_archivos_train_zip(train_zip_file):
    nombres_archivos = []  # Lista para almacenar los nombres de los archivos
    with zipfile.ZipFile(train_zip_file, 'r') as train_zip_ref:
        for archivo in train_zip_ref.namelist():
            if archivo.startswith('train/'):
                nombre_archivo = os.path.basename(archivo)  # Obtener solo el nombre del archivo sin la ruta
                nombres_archivos.append(nombre_archivo)  # Agregar nombre de archivo a la lista
    nombres_archivos.sort()  # Ordenar alfanuméricamente
    for nombre_archivo in nombres_archivos:
        print(nombre_archivo)  # Imprimir nombre del archivo
    return nombres_archivos

# Uso:
ruta_zip = '/content/drive/MyDrive/Datos/proyecto_redes_neuronales/dogs-vs-cats.zip'
listar_archivos_zip(ruta_zip)
```

```python
import zipfile

def contar_archivos_jpg(ruta_zip):
    total_jpg = 0
    with zipfile.ZipFile(ruta_zip, 'r') as zip_ref:
        for file_info in zip_ref.infolist():
            if file_info.filename.endswith('train.zip'):
                with zip_ref.open(file_info) as train_zip_file:
                    with zipfile.ZipFile(train_zip_file, 'r') as train_zip_ref:
                        lista_archivos = train_zip_ref.namelist()
                        total_jpg += sum(1 for archivo in lista_archivos if archivo.endswith('.jpg'))
    return total_jpg

# Llamar a la función con la ruta del archivo zip
ruta_zip = '/content/drive/MyDrive/Datos/proyecto_redes_neuronales/dogs-vs-cats.zip'
total_jpg = contar_archivos_jpg(ruta_zip)
print(f"Total de archivos .jpg dentro de train.zip: {total_jpg}")
```

```python
import zipfile
import os
import matplotlib.pyplot as plt
from PIL import Image

def descargar_imagenes(ruta_zip, destino):
    # Crear la carpeta de destino si no existe
    if not os.path.exists(destino):
        os.makedirs(destino)

    with zipfile.ZipFile(ruta_zip, 'r') as zip_ref:
        # Buscar el archivo train.zip dentro de dogs-vs-cats.zip
        for file_info in zip_ref.infolist():
            if file_info.filename.endswith('train.zip'):
                # Extraer el contenido de train.zip en una carpeta temporal
                with zip_ref.open(file_info) as train_zip_file:
                    with zipfile.ZipFile(train_zip_file, 'r') as train_zip_ref:
                        # Extraer las imágenes de la carpeta train dentro de train.zip
                        for i in range(9):
                            nombre_archivo_cat = f'train/cat.{i}.jpg'
                            nombre_archivo_dog = f'train/dog.{i}.jpg'
                            try:
                                img_cat = train_zip_ref.read(nombre_archivo_cat)
                                img_dog = train_zip_ref.read(nombre_archivo_dog)
                                ruta_destino_cat = os.path.join(destino, f'cat.{i}.jpg')
                                ruta_destino_dog = os.path.join(destino, f'dog.{i}.jpg')
                                with open(ruta_destino_cat, 'wb') as file_cat:
                                    file_cat.write(img_cat)
                                print(f"Imagen guardada en: {ruta_destino_cat}")
                                with open(ruta_destino_dog, 'wb') as file_dog:
                                    file_dog.write(img_dog)
                                print(f"Imagen guardada en: {ruta_destino_dog}")
                            except KeyError as e:
                                print(f"No se pudo encontrar el archivo: {e}")
                break

# Uso:
ruta_zip = '/content/drive/MyDrive/Datos/proyecto_redes_neuronales/dogs-vs-cats.zip'
destino = '/content/images'
descargar_imagenes(ruta_zip, destino)

```

```python
import io

import zipfile
import io
from PIL import Image

def listar_archivos_train_zip(ruta_zip):
    with zipfile.ZipFile(ruta_zip, 'r') as zip_ref:
        # Buscar el archivo train.zip dentro de dogs-vs-cats.zip
        for file_info in zip_ref.infolist():
            if file_info.filename.endswith('train.zip'):
                # Extraer el contenido de train.zip en una carpeta temporal
                with zip_ref.open(file_info) as train_zip_file:
                    with zipfile.ZipFile(train_zip_file, 'r') as train_zip_ref:
                        archivos = train_zip_ref.namelist()
                        for archivo in archivos:
                            print(f"   {archivo}")
                            if archivo.lower().endswith('.jpg'):
                                with train_zip_ref.open(archivo) as file:
                                    imprimir_tamaño(file, archivo)

def imprimir_tamaño(archivo_zip, nombre_archivo):
    with io.BytesIO(archivo_zip.read()) as file_buffer:
        img = Image.open(file_buffer)
        tamaño = img.size
        print(f"Tamaño de {nombre_archivo}: {tamaño[0]} x {tamaño[1]} píxeles")

# Uso:
ruta_zip = '/content/drive/MyDrive/Datos/proyecto_redes_neuronales/dogs-vs-cats.zip'
listar_archivos_train_zip(ruta_zip)
```

```python
import zipfile

def calcular_tamaño_total_en_zip(ruta_zip, carpeta_train):
    tamaño_total = 0
    with zipfile.ZipFile(ruta_zip, 'r') as zip_ref:
        for file_info in zip_ref.infolist():
            if file_info.filename.endswith('train.zip'):
                with zip_ref.open(file_info) as train_zip_file:
                    with zipfile.ZipFile(train_zip_file, 'r') as train_zip_ref:
                        for archivo_train in train_zip_ref.infolist():
                            if archivo_train.filename.startswith(carpeta_train):
                                tamaño_total += archivo_train.file_size
                break
    return tamaño_total

# Uso:
ruta_zip = '/content/drive/MyDrive/Datos/proyecto_redes_neuronales/dogs-vs-cats.zip'
carpeta_train = 'train/'
tamaño_total = calcular_tamaño_total_en_zip(ruta_zip, carpeta_train)
print(f"El tamaño total de las imágenes en la carpeta '{carpeta_train}' es de {tamaño_total} bytes")
```

```python
import zipfile
import io
from PIL import Image
import os

import zipfile
import os
import io
from PIL import Image

def redimensionar_y_guardar_imagenes_desde_zip(ruta_zip, carpeta_principal):
    # Crear las carpetas si no existen
    carpeta_destino = os.path.join(carpeta_principal, 'train')
    os.makedirs(carpeta_destino, exist_ok=True)

    with zipfile.ZipFile(ruta_zip, 'r') as zip_ref:
        for archivo in sorted(zip_ref.namelist()):  # Ordenar alfabéticamente los nombres de archivo
            if archivo.endswith('train.zip'):
                with zip_ref.open(archivo) as train_zip_file:
                    train_zip_content = io.BytesIO(train_zip_file.read())
                    with zipfile.ZipFile(train_zip_content, 'r') as train_zip_ref:
                        for archivo_train in sorted(train_zip_ref.namelist()):  # Ordenar alfabéticamente los nombres de archivo
                            if archivo_train.endswith('.jpg'):
                                with train_zip_ref.open(archivo_train) as file:
                                    img = Image.open(io.BytesIO(file.read()))
                                    tamaño_original = img.size
                                    print(f"Tamaño original de la imagen {archivo_train}: {tamaño_original[0]} x {tamaño_original[1]} píxeles")

                                    # Redimensionar la imagen
                                    img_redimensionada = img.resize((200, 200))
                                    tamaño_redimensionado = img_redimensionada.size
                                    print(f"Tamaño redimensionado de la imagen {archivo_train}: {tamaño_redimensionado[0]} x {tamaño_redimensionado[1]} píxeles")

                                    # Guardar la imagen redimensionada
                                    ruta_guardado = os.path.join(carpeta_destino, os.path.basename(archivo_train))
                                    img_redimensionada.save(ruta_guardado)

                                    print(f"Imagen guardada en: {ruta_guardado}")

# Uso:
ruta_zip = '/content/drive/MyDrive/Datos/proyecto_redes_neuronales/dogs-vs-cats.zip'
carpeta_principal = '/content/dataset'
redimensionar_y_guardar_imagenes_desde_zip(ruta_zip, carpeta_principal)

```

```python
import os
import pandas as pd

# Rutas de las carpetas de train
ruta_cat = '/content/dataset/train/cat'
ruta_dog = '/content/dataset/train/dog'

# Obtener la lista de archivos en las carpetas de train
archivos_cat = os.listdir(ruta_cat)
archivos_dog = os.listdir(ruta_dog)

# Asignar ID a los archivos de la carpeta cat
ids_cat = list(range(len(archivos_cat)))
ids_cat = [str(id) for id in ids_cat]

# Asignar ID a los archivos de la carpeta dog
ids_dog = list(range(len(archivos_dog)))
ids_dog = [str(id) for id in ids_dog]

# Crear DataFrame para train con los IDs de los archivos y sus respectivas etiquetas
df_train_cat = pd.DataFrame({'id': ids_cat, 'label': 0})  # Asignar label 0 para cat
df_train_dog = pd.DataFrame({'id': ids_dog, 'label': 1})  # Asignar label 1 para dog

# Concatenar los DataFrames de cat y dog
df_train = pd.concat([df_train_cat, df_train_dog], ignore_index=True)

# Alternar las etiquetas (labels) entre 0 y 1
df_train['label'] = df_train['label'].astype(str)
df_train['label'] = df_train.groupby('label').cumcount() % 2

# Verificar el DataFrame resultante
df_train.head()
```

INICIALIZO EL MODELO

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Definir el generador de datos para el conjunto de entrenamiento
train_datagen = ImageDataGenerator(
    rescale=1.0 / 255.0,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)
```

```python
# Crear generador de flujo de datos para el conjunto de entrenamiento
train_generator = train_datagen.flow_from_directory(
    '/content/dataset/train',
    target_size=(224, 224),  # Tamaño de las imágenes de entrada para la red VGG16
    batch_size=32,
    class_mode='binary'
)

```

```python
test_datagen = ImageDataGenerator(rescale=1.0 / 255.0)

test_generator = test_datagen.flow_from_directory(
    '/content/dataset',
    classes=['test'],  # Esto asegura que solo lea las imágenes en la carpeta 'test'
    target_size=(224, 224),  # Tamaño de entrada para VGG16
    batch_size=32,
    class_mode='binary',  # No se utilizan etiquetas
    shuffle=False
)
```

```python
# Definir el modelo
model = Sequential([
    Conv2D(input_shape=(224, 224, 3), filters=32, kernel_size=(3, 3), padding="same", activation="relu"),
    MaxPooling2D(pool_size=(2, 2), strides=(2, 2)),
    Conv2D(filters=64, kernel_size=(3, 3), padding="same", activation="relu"),
    MaxPooling2D(pool_size=(2, 2), strides=(2, 2)),
    Conv2D(filters=128, kernel_size=(3, 3), padding="same", activation="relu"),
    MaxPooling2D(pool_size=(2, 2), strides=(2, 2)),
    Flatten(),
    Dense(units=2048, activation="relu"),
    Dense(units=2, activation="softmax")
])
```

```python
model.compile(optimizer='SGD', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(train_generator, epochs=5, validation_data=test_generator)
```
